# Tugas 1 : Data Integration

```python
import pandas as pd

#MENGGABUNGKAN DATA DENGAN "MERGE"

data_salesM1 = pd.read_csv("D:\DATA MINING\Data_SalesM1.csv")
data_salesM2 = pd.read_csv("D:\DATA MINING\Data_SalesM2.csv")

data_salesM = pd.merge(data_salesM1, data_salesM2, on=["Product_Code"])
data_salesM.head()
```

```python
import pandas as pd

#MENGGABUNGKAN DATA DENGAN "CONCAT"

data_salesC1 = pd.read_csv("D:\DATA MINING\Data_SalesC1.csv")
data_salesC2 = pd.read_csv("D:\DATA MINING\Data_SalesC2.csv")

data_salesC = pd.concat([data_salesC1, data_salesC2])
data_salesC.head()
```

```python
data_salesM.to_csv("datasales1.csv")
data_salesC.to_csv("datasales2.csv")
```
