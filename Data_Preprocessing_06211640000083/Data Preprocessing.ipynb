```python
import pandas as pd
import numpy as np
```

```python
data_movie=pd.read_csv("D:\DATA MINING\movie_metadata.csv")
```

```python
#Command untuk statdes di Phyton

data_movie.describe() 
```

```python
#Untuk mengetahui missing value (NA)

np.sum(data_movie.isnull())
```

```python
#Jika missing value terdapat kurang dari 10%, maka boleh dihapus
#Jika missing value lebih dari 10%, maka dilakukan dengan imputasi. Jika datanya kontinyu, bisa diisi dengan rata-rata
```

```python
#Untuk mengisi NA dengan nilai rata-rata pada variabel "budget"
data_movie["budget"]=data_movie["budget"].fillna(3.975262e+07)

#Untuk mengecek apakah NA pada variabel "budget" sudah teratasi
np.sum(data_movie.isnull())
```

```python
data_movie["gross"]=data_movie["gross"].fillna(4.846841e+07)

np.sum(data_movie.isnull())
```

```python
#PER-VARIABEL

#1. Kurang dr 10%, diabaikan atau diimputasi. 
#2. Lebih dari 15%, bisa jadi kandidat yg akan dihapus (jika punya korelasi besar, diimputasi).
#3. Lebih dari 50%, maka variabelnya yang dihapus.

#PER-PENGAMATAN (pelajari di buku Hair)
```

```python
#Fill Missing Value with Mean

data_movie["duration"]=data_movie["duration"].fillna((data_movie["budget"].mean()))
data_movie["facenumber_in_poster"]=data_movie["facenumber_in_poster"].fillna((data_movie["facenumber_in_poster"].mean()))
data_movie["num_user_for_reviews"]=data_movie["num_user_for_reviews"].fillna((data_movie["num_user_for_reviews"].mean()))
data_movie["num_critic_for_reviews"]=data_movie["num_critic_for_reviews"].fillna((data_movie["num_critic_for_reviews"].mean()))
data_movie["director_facebook_likes"]=data_movie["director_facebook_likes"].fillna((data_movie["director_facebook_likes"].mean()))
data_movie["actor_1_facebook_likes"]=data_movie["actor_1_facebook_likes"].fillna(6561.323932)
data_movie["actor_2_facebook_likes"]=data_movie["actor_2_facebook_likes"].fillna(1652.080533)
data_movie["actor_3_facebook_likes"]=data_movie["actor_3_facebook_likes"].fillna(645.009761)

np.sum(data_movie.isnull())
```

```python
#Kapan imputasi pake mean atau median?

#1. Lihat distribusi data 
#Jika tidak berdistribusi normal, atau mempunyai skewness positif, negatif, memakai median.
#Jika data berdistribusi normal, memakai mean
```

```python
#Untuk mengisi missing value yang datanya berbentuk kategorik

from scipy.stats import mode

data_movie["country"]=data_movie["country"].fillna(data_movie["country"]).dropna().mode()[0]
np.sum(data_movie.isnull())
```

```python
#Mengatasi Outlier

from scipy import stats
z = np.abs(stats.zscore(data_movie._get_numeric_data()))
print(z)
data_movie = data_movie[(z<3).all(axis=1)]
print(data_movie.shape)
```

```python
#Outlier

#Univariat : jika pengamatan 1, variabel 1, outlier, maka dihapus sebaris.
#Multivariat : jika pengamatan 1, dari keseluruhan variabel (secara simultan) outlier, dihapus sebaris.
```
