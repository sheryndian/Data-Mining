# Data Pre-processing

## Melbourne Airbnb Open Data
## Source : Kaggle

```python
import numpy as np
import pandas as pd
```

```python
airbnb = pd.read_csv("D:\AirbnbMelb.csv")
airbnb
```

```python
#Mengidentifikasi Missing Value

np.sum(airbnb.isnull())
```

```python
# Statdes pada data Airbnb

airbnb.describe()
```

```python
# Filling Missing Value with Mean

airbnb["weekly_price"]=airbnb["weekly_price"].fillna((airbnb["weekly_price"].mean()))
np.sum(airbnb.isnull())
```

```python
# Filling Missing Value with Mean

airbnb["monthly_price"]=airbnb["monthly_price"].fillna((airbnb["monthly_price"].mean()))
airbnb["security_deposit"]=airbnb["security_deposit"].fillna((airbnb["security_deposit"].mean()))
airbnb["cleaning_fee"]=airbnb["cleaning_fee"].fillna((airbnb["cleaning_fee"].mean()))
airbnb["review_scores_rating"]=airbnb["review_scores_rating"].fillna((airbnb["review_scores_rating"].mean()))
np.sum(airbnb.isnull())
```

```python
airbnb
```

```python
# Mengatasi Outlier

from scipy import stats
z = np.abs(stats.zscore(airbnb._get_numeric_data()))
print(z)
airbnb= airbnb[(z<3).all(axis=1)]
print(airbnb.shape)
```
