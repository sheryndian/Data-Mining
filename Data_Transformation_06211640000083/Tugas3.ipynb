```python
import pandas as pd
import numpy as np
```

```python
shanghai = pd.read_csv("D:\DATA MINING\shanghai.csv")
shanghai.tail()
```

# 1. Missing Value

```python
#Identifikasi Missing Value
np.sum(shanghai.isnull())
```

```python
shanghai.describe()
```

```python
#Filling Missing Value

shanghai['PM_Jingan']=shanghai['PM_Jingan'].fillna(shanghai['PM_Jingan'].mean())
shanghai['DEWP']=shanghai['DEWP'].fillna(shanghai['DEWP'].mean())
shanghai['HUMI']=shanghai['HUMI'].fillna(shanghai['HUMI'].mean())
shanghai['PRES']=shanghai['PRES'].fillna(shanghai['PRES'].mean())
shanghai['TEMP']=shanghai['TEMP'].fillna(shanghai['TEMP'].mean())
shanghai['Iws']=shanghai['Iws'].fillna(shanghai['Iws'].mean())
shanghai['precipitation']=shanghai['precipitation'].fillna(shanghai['precipitation'].mean())
shanghai['Iprec']=shanghai['Iprec'].fillna(shanghai['Iprec'].mean())

np.sum(shanghai.isnull())
```

# 2. Missing Value, Outlier

```python
from scipy.stats import stats
z= np.abs(stats.zscore(shanghai._get_numeric_data()))
print(z)
shanghai2=shanghai[(z<3).all(axis=1)]
print(shanghai2.shape)
```

```python
x2=shanghai2.iloc[:,6:11]
y2 = shanghai2.iloc[:,5]
model2 = stat.ols("y2 ~ x2" , data=shanghai2).fit()
print(model2.summary())
```

# 3. Missing Value, Transformasi

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-10,10))
cut_shanghai= shanghai.iloc[:,5:11]
scaler.fit(cut_shanghai)
x_scaled = scaler.transform(cut_shanghai)
x_scaled= pd.DataFrame(x_scaled)
x_scaled.tail()
```

```python
x3=cut_shanghai.iloc[:,1:6]
y3 = cut_shanghai.iloc[:,0]
model3 = stat.ols("y3 ~ x3" , data=cut_shanghai).fit()
print(model3.summary())
```
