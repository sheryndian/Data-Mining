```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
ets5 = pd.read_csv("ets5.csv", engine='python')
ets5.head()
```

```python
ets5.shape
```

```python
np.sum(ets5.isnull())
```

```python
for i, j in ets5[ets5['Description'].isnull()].iterrows():ets5['Description'][i]="Code-"+str(j["StockCode"])
```

```python
np.sum(ets5.isnull())
```

```python
a = ets5.groupby("InvoiceNo")
purchase=[]
for i,j in a:
    purchase.append(list(j["Description"].map(str)))
```

```python
x = pd.DataFrame(purchase)
```

```python
x
```

```python
association_rules=apriori(purchase, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result :
    #first index of the inner list
    #contains base item and add item
    pair = item [0]
    items = [a for a in pair]
    print("rule: " + items[0] + " ->" + items[1])
    
    #second index of the inner list
    print("support: " + str(item[1]))
    
    #third index of the list located at 10th
    #of the third index of the inner list
    
    print("confidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("==========================================================================")
```
