```python
AUS["MinTemp"]=AUS["MinTemp"].fillna((AUS["MinTemp"].mean()))
AUS["MaxTemp"]=AUS["MaxTemp"].fillna((AUS["MaxTemp"].mean()))
AUS["Rainfall"]=AUS["Rainfall"].fillna((AUS["Rainfall"].mean()))
AUS["Evaporation"]=AUS["Evaporation"].fillna((AUS["Evaporation"].mean()))
AUS["Sunshine"]=AUS["Sunshine"].fillna((AUS["Sunshine"].mean()))
AUS["WindGustSpeed"]=AUS["WindGustSpeed"].fillna((AUS["WindGustSpeed"].mean()))
AUS["WindSpeed9am"]=AUS["WindSpeed9am"].fillna((AUS["WindSpeed9am"].mean()))
AUS["WindSpeed3pm"]=AUS["WindSpeed3pm"].fillna((AUS["WindSpeed3pm"].mean()))
AUS["Humidity9am"]=AUS["Humidity9am"].fillna((AUS["Humidity9am"].mean()))
AUS["Humidity3pm"]=AUS["Humidity3pm"].fillna((AUS["Humidity3pm"].mean()))
AUS["Pressure9am"]=AUS["Pressure9am"].fillna((AUS["Pressure9am"].mean()))
AUS["Pressure3pm"]=AUS["Pressure3pm"].fillna((AUS["Pressure3pm"].mean()))
AUS["Cloud9am"]=AUS["Cloud9am"].fillna((AUS["Cloud9am"].mean()))
AUS["Cloud3pm"]=AUS["Cloud3pm"].fillna((AUS["Cloud3pm"].mean()))
AUS["Temp9am"]=AUS["Temp9am"].fillna((AUS["Temp9am"].mean()))
AUS["Temp3pm"]=AUS["Temp3pm"].fillna((AUS["Temp3pm"].mean()))
AUS["RISK_MM"]=AUS["RISK_MM"].fillna((AUS["RISK_MM"].mean()))
```

```python
np.sum(AUS.isnull())
```

```python
from scipy.stats import mode
AUS['WindGustDir']=AUS['WindGustDir'].fillna(AUS['WindGustDir']).dropna().mode()[0]
AUS['WindDir9am']=AUS['WindDir9am'].fillna(AUS['WindDir9am']).dropna().mode()[0]
AUS['WindDir3pm']=AUS['WindDir3pm'].fillna(AUS['WindDir3pm']).dropna().mode()[0]
AUS['RainToday']=AUS['RainToday'].fillna(AUS['RainToday']).dropna().mode()[0]
AUS['RainTomorrow']=AUS['RainTomorrow'].fillna(AUS['RainTomorrow']).dropna().mode()[0]

np.sum(AUS.isnull())
```

```python
#Mengatasi Outlier

from scipy import stats
z = np.abs(stats.zscore(AUS._get_numeric_data()))
print(z)
AUS = AUS[(z<3).all(axis=1)]
print(AUS.shape)
```

```python
AUS
```

```python
AUS1=AUS.drop(columns=['Date','Location','WindGustDir','WindDir9am','WindDir3pm'])
AUS1['RainToday'].replace({'No':0,'Yes':1},inplace=True)
AUS1['RainTomorrow'].replace({'No':0,'Yes':1},inplace=True)
AUS1.tail()
```

```python
from sklearn.preprocessing import normalize
xnorm=normalize(AUS1,norm='l2') 
xnorm=pd.DataFrame(xnorm) 
xnorm.tail()
```

```python
from sklearn.feature_selection import SelectKBest, chi2
x = AUS.drop(columns=['RainTomorrow'])
y = AUS.[['RainTomorrow']]
selector = SelectKBest(chi2, k='all').fit(x,y)
x_new=selector.transform(x)
scores=selector.scores_
```
